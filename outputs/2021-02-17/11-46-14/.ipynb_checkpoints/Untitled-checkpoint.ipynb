```python
import os
import ast
import numpy as np
import pandas as pd
```

```python
filenames = os.listdir("./")
filenames = [filename for filename in filenames if filename.endswith(".csv") and filename.startswith("archive")]
```

```python
li = []
for filename in filenames:
    df = pd.read_csv(filename, header=0, index_col=False)
    df["generation"] = filename.split("_")[1].split(".")[0]
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame
```

```python
frame.pedigree = frame.pedigree.apply(ast.literal_eval)
frame.generation = frame.generation.apply(ast.literal_eval)
```

```python
frame["origin"] = frame["pedigree"].apply(lambda x: x[0])
frame["reaction / parent"] = frame["pedigree"].apply(lambda x: x[1])
frame["parent"] = frame["pedigree"].apply(lambda x: x[2])
```

```python
frame.drop(columns=["pedigree"], axis=0, inplace = True)
```

```python
frame.sort_values(by=['generation'],inplace=True)
```

```python
frame["reward"] = 0.0
for index, row in frame.iterrows():
    subset = frame[frame["index"] == row["index"]]
    subset = subset[subset["generation"] < float(row["generation"])]
    previous = subset[subset.generation == subset.generation.max()]
    previous_value = previous.fitnesses.values
    if previous_value.size > 0:
        increase = row.fitnesses - previous_value[0]
        frame.at[index, 'reward'] = increase
    else:
        frame.at[index, 'reward'] = row.fitnesses 
```

```python
frame.drop_duplicates(subset=['smiles'], inplace=True)
```

```python
frame.reset_index(drop=True, inplace=True)
frame
```

```python
frame.groupby(['origin', 'generation'])[["fitnesses"]].max().to_csv("summary.csv")
```

```python
frame.groupby(['origin'])[["fitnesses", "reward"]].max()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 

sns.set_theme(style="whitegrid")
%matplotlib inline
plt.figure(figsize=(200, 800))
g = sns.relplot(data=frame, x="generation", y="fitnesses", hue="origin", size="reward")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
plt.savefig("figure", dpi=500)
```

```python
sns.set_theme(style="whitegrid")
g = sns.relplot( data=frame, x="generation", y="fitnesses", hue="origin", size="reward", palette="muted", height=15, sizes=(50, 250),)

g.despine(left=True, bottom=True)
g.ax.margins(.02)
for label in g.ax.get_xticklabels():
    label.set_rotation(90)
for artist in g.legend.legendHandles:
    artist.set_edgecolor(".7")
    g.ax.xaxis.grid(True, "minor", linewidth=.25)
plt.savefig("figure", dpi=500)
```

```python
sns.set_theme(style="whitegrid")
g = sns.relplot( data=frame, x="index", y="fitnesses", hue="origin", size="reward", palette="muted", height=15, sizes=(50, 250),)

g.despine(left=True, bottom=True)
g.ax.margins(.02)
for label in g.ax.get_xticklabels():
    label.set_rotation(90)
for artist in g.legend.legendHandles:
    artist.set_edgecolor(".7")
    g.ax.xaxis.grid(True, "minor", linewidth=.25)
plt.tight_layout()
plt.savefig("figure", dpi=500, pad_inches=0.5)
```
