```python
import os

import numpy as np
import pandas as pd
```

```python
filenames = os.listdir("./")
filenames = [filename for filename in filenames if filename.endswith(".csv") and filename.startswith("archive")]
```

```python
li = []
for filename in filenames:
    df = pd.read_csv(filename, header=0, index_col=False)
    df["generation"] = filename.split("_")[1].split(".")[0]
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
frame.pedegree = frame.pedegree.apply(ast.literal_eval)
```

```python
frame["origin"] = frame["pedegree"].apply(lambda x: x[0])
frame["reaction / parent"] = frame["pedegree"].apply(lambda x: x[1])
frame["parent"] = frame["pedegree"].apply(lambda x: x[2])
```

```python
frame.drop(columns=["index", "pedegree"], axis=0, inplace = True)
```

```python
frame["reward"] = 0.0
```

```python
for index, row in frame.iterrows():
    if row["origin"] == "mutation":
        frame.at[index, "reward"] = row["fitnesses"] - np.mean(frame[frame["smiles"] == row["parent"]]["fitnesses"].values)
    if row["origin"] == "crossover":
        value_1 = np.mean(frame[frame["smiles"] == row["parent"]]["fitnesses"].values)
        value_2 = np.mean(frame[frame["smiles"] == row["reaction / parent"]]["fitnesses"].values)
        value = np.max([value_1, value_2])
        frame.at[index, "reward"] = row["fitnesses"] - value
```
